```python
import os
import glob
import cv2
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Paths
image_dir = r"C:\Users\LENOVO\Downloads\Programs\New folder (3)\New folder"
label_dir = r"C:\Users\LENOVO\Downloads\Programs\New folder (3)\New folder"

# Get all image files
image_files = glob.glob(os.path.join(image_dir, "*.jpg"))

# Class names
class_names = ["CANCER CELL"]

# Setup matplotlib figure
plt.ion()  # Interactive mode on

for image_path in image_files:
    # Read image with OpenCV (BGR format)
    img = cv2.imread(image_path)
    if img is None:
        print(f"Could not read image: {image_path}")
        continue
    
    # Convert to RGB for matplotlib
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # Create figure
    fig, ax = plt.subplots(1)
    ax.imshow(img_rgb)
    
    # Get corresponding label file
    base_name = os.path.splitext(os.path.basename(image_path))[0]
    label_path = os.path.join(label_dir, base_name + ".txt")
    
    if os.path.exists(label_path):
        with open(label_path, 'r') as f:
            lines = f.readlines()
        
        img_height, img_width = img.shape[:2]
        
        for line in lines:
            parts = line.strip().split()
            if len(parts) != 5:
                continue
                
            class_id, x_center, y_center, w, h = map(float, parts)
            class_id = int(class_id)
            
            # Convert YOLO to pixel coordinates
            x_min = (x_center - w/2) * img_width
            y_min = (y_center - h/2) * img_height
            box_width = w * img_width
            box_height = h * img_height
            
            # Create rectangle patch
            rect = patches.Rectangle(
                (x_min, y_min), box_width, box_height,
                linewidth=1, edgecolor='lime', facecolor='none'
            )
            ax.add_patch(rect)
            
            # Add label
            plt.text(
                x_min, y_min - 5, class_names[class_id],
                color='lime', fontsize=8,
                bbox=dict(facecolor='black', alpha=0.5, edgecolor='none')
            )
    
    plt.title(os.path.basename(image_path))
    plt.axis('off')
    plt.show()
    plt.pause(0.5)  # Pause to view the image
    input("Press Enter to continue...")  # Wait for user input
    plt.close()

plt.ioff()
```
